```python
# subset DataFrame with original file columns and whether the reaction stopped when drug therapy was stopped (dechal) and if the reaction
# began again after drug therapy was restarted (rechal). Append into list for merging with class_dfs.
additions = []
custom_dfs = []
def file_merge(saved_dfs, drug_files, final_reacs, final_outs):
    
    for sf,df in zip(saved_dfs, drug_files):
        indices = f.orig_idx
        kept = df[['primaryid','caseid']].loc[indices]
        additions.append(kept)
        
    for f,a,rf,of in zip(saved_dfs, additions, final_reacs, final_outs):
        new = f.join([a.set_index(f.index)])
        new['pt'] = 'nan'
        new['outc_cod'] = 'nan'
        i = 0

        for i,x in enumerate(rf.primaryid):
            for j,y in enumerate(new.primaryid):
                if x == y:
                    new.pt.loc[j] = rf.pt.loc[i]
        for i,x in enumerate(of.primaryid):
            for j,y in enumerate(new.primaryid):
                if x == y:
                    new.outc_cod.loc[j] = of.outc_cod.loc[i]


    custom_dfs.append(new)
```

```python
# for loop for determining the amount of 'PS' (Primary suspect), 'SS' (Secondary Suspects), 'C', (concamitants), 
# or 'I' (Interacting) medications present for every unique CaseID ***6.5hr runtime***

start_time = time.time()
p = 0
s = 0
c = 0
i = 0


for x,j in enumerate(new_drug_df.primaryid.unique()):
    df = new_drug_df[new_drug_df.primaryid == j]


    for y in range(len(df)):
        if df.iloc[y,1] == 'PS':
            p += 1
        elif df.iloc[y,1] == 'SS':
            s += 1
        elif df.iloc[y,1] == 'c':
            c += 1
        else:
            i += 1

    drugs.loc[x, 'PrimaryID'] = j
    drugs.loc[x, 'Primary'] = p
    drugs.loc[x, 'Secondary'] = s
    drugs.loc[x, 'Concamitant'] = c
    drugs.loc[x, 'Interacting'] = i
        
    p,s,c,i = 0,0,0,0
end_time = time.time()

total_time = (end_time - start_time) / 60
total_time / 60
```
